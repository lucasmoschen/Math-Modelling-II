```python
import matplotlib.pyplot as plt
```

```python
def plota(string):
    f = open(string,'r')
    lista = []
    for r in f.readlines():
        lista.append(float(r))
    plt.plot(lista)
    plt.show()
```

```python
plota('analise_energias_k0_r1.txt')
```

```python
plota('analise_energias_nao_conserva.txt')
```
